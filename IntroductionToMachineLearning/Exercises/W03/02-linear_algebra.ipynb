**Task 3 (easy): Determine weights 1**
Given the affine model: 
\[y = w_1 x_1 + w_2 x_2 + w_3\]
Determine which of the weight options $(w_1, w_2, w_3)$ contain all data points below, such that all three points are on the plane:
**Data points:**
Input $x_i$, label $y$:
- $x_1 = 2, x_2 = 0, y = 5  $
- $x_1 = 1, x_2 = 1, y= 6  $
- $x_1 = 0, x_2 = 2, y= 7  $

**Options:**
1. $(w_1, w_2, w_3) = (1, 0, 3)$
2. $(w_1, w_2, w_3) = (1, 3, 2)$
3. $(w_1, w_2, w_3) = (2, 3, 1)$


---

```python
#Write your answer here...
def opt1(pnt):
    x1, x2, y = pnt
    return 1 * x1 + 0 * x2 + 3 == y
def opt2(pnt):
    x1, x2, y = pnt
    return 1 * x1 + 3 * x2 + 2 == y
def opt3(pnt):
    x1, x2, y = pnt
    return 2 * x1 + 3 * x2 + 1 == y

points = [
    [2, 0, 5],
    [1, 1, 6],
    [0, 2, 7]
]
points_opt1 = [opt1(p) for p in points]
points_opt2 = [opt2(p) for p in points]
points_opt3 = [opt3(p) for p in points]

print(f"Opt1: {points_opt1}")
print(f"Opt2: {points_opt2}")
print(f"Opt3: {points_opt3}") # The correct set of weights is option 3 with w1 = 0, w2 = 2 & w3 = 7
```

**Task 4 (medium): Determine weights 2**
Given the affine model:
\[y = w_1 x_1 + w_2 x_2 + w_3\]
1. Determine the values of the weights $w = (w_1, w_2, w_3)$ that make the model fit all of the points:

- $x_1 =1, x_2= 0, y = 2$
- $x_1 = 0, x_2 = 1, y = 3$
- $x_1 = 1, x_2 = 1, y = 6$

2. Rewrite the affine model as an inner-product using inputs and labels.
3. Rewrite the model in matrix form $(Aw = y)$ using the input points and labels, with $w$ as the vector of unknowns to solve for.


---

```python
#Write your answer here
import numpy as np
A = np.array([
    [1, 0, 1],
    [0, 1, 1],
    [1, 1, 1]
])

y = np.array([2, 3, 6])

# Solve for weights
w = np.linalg.solve(A, y) # A^{-1} @ y

print(f"w1 = {w[0]}")
print(f"w2 = {w[1]}")
print(f"w3 = {w[2]}")
print(f"\ny = {w[0]} * x1 + {w[1]} * x2 + {w[2]}")
```
